```julia
using Plots; plotlyjs();    # PlotlyJS backend
using DensityInterpolationMethod
using DensityInterpolationMethod.Mesh
using DensityInterpolationMethod.Utils 
using DensityInterpolationMethod.IO
```

```julia
# Read mesh
ELEM_ORDER = 1    # element order
HMAX = 1          # max mesh size
mesh_filename = "../test/meshes/sphere1.geo"

mesh = read_gmsh_geo(mesh_filename, h=HMAX, order=ELEM_ORDER);
```

# Mesh node plot

```julia
meshplot(mesh, centers=false, normals=0.5, tangents=0, fmt = :svg)
```

# Mesh surface plot
NOTE: It is possible that some elements may not be displayed properly.

```julia
surfacemeshplot(mesh, npoints=10, fmt = :svg)
meshplot!(mesh, nodes=false, normals=0.5)
```

# Plots also work on quadratic elements

```julia
# Read mesh
ELEM_ORDER = 2    # element order
mesh = read_gmsh_geo(mesh_filename, h=HMAX, order=ELEM_ORDER);
```

```julia
meshplot(mesh, centers=false, normals=0.5, tangents=0, fmt = :svg)
```
